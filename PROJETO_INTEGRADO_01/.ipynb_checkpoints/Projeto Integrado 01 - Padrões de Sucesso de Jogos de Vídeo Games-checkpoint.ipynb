3.2.1 Coluna <b>name</b>.

Vamos verificar os valores ausentes na coluna <b>name</b>.
    </font>

```python
# Exibindo valores ausentes
df[df['name'].isna()]
```

<font face='Calibri' color="#001A56" size=+0.8>
Veremos a porcentagem que essas duas únicas entradas representam de todo o dataset.

```python
df[df['name'].isna()].count()/len(df['name'])
```

<font face='Calibri' color="#001A56" size=+0.8>
Como não é possível inferir o nome do jogo que está ausente com base nas demais informações e nesses mesmos dados temos outras informações também ausentes, iremos excluí-los, pois entendemos que a quantidade de dados que serão excluídos não afetará as análises posteriores.

```python
# Excluindo os dados
df = df.dropna(subset=['name']).reset_index(drop=True)

# Exibindo informações
df.info()
```

<font face='Calibri' color="#001A56" size=+0.8>
Os valores ausentes na coluna <b>name</b> foram excluídos, com essas exclusões também foram eliminados os valores ausentes na coluna <b>genre</b>.

<a id='3.2.2'></a>
<h4><font face='Calibri' color="#1E90B6" size=+1.8>
3.2.2 Coluna <b>year_of_release</b>.
</font></h4>
    <p>
<font face='Calibri' color="#001A56" size=+0.8>
Trataremos agora da coluna  <b>year_of_release</b>, começaremos verificando os valores ausentes.

```python
# Contando valores
df['year_of_release'].value_counts(dropna=False)
```

<font face='Calibri' color="#001A56" size=+0.8>
Temos 269 jogos com valores ausentes em  <b>year_of_release</b>. Será que esses valores ausentes tem alguma relação com o gênero ou plataforma? iremos checar isso à seguir para que possamos substituir os valores ausentes.

```python
# Agrupando por gêneros.
df.pivot_table(index='genre', values='year_of_release', aggfunc=['mean','median'])
```

<font face='Calibri' color="#001A56" size=+0.8>
Se agruparmos os jogos por gênero e obtivermos a média ou mediana desses jogos vamos obter valores muito parecidos para todos os jogos, além de que seria um erro assumirmos que todos jogos de ação, por exemplo, foram lançados em 2009 pois sabemos que temos jogos muito mais antigos que isso (Doom, por exemplo). Por esses motivos não usaremos a média ou mediana por gênero para preencher valores ausentes.

```python
# Agrupando por plataforma
df.pivot_table(index='platform', values='year_of_release', aggfunc=['mean','median']).head(30)
```

<font face='Calibri' color="#001A56" size=+0.8>
Notamos que o ano de lançamento tem maior relação com a plataforma do que com o gênero, o que faz sentido pois novas plataformas são lançadas conforme o avanço da tecnologia. Utilizaremos a mediana dos anos de lançamento agrupados por plataforma para preencher os valores ausentes. Será utilizado a mediana para que tenhamos menos números quebrados, para evitar a distorsão de possíveis valores atípicos e por não ter muita diferença em relação á média.

```python
# Preenchendo os valores
df['year_of_release'].fillna(df.groupby('platform')['year_of_release'].transform('median'), inplace=True)

# Exibindo informações
df.info()
```

<font face='Calibri' color="#001A56" size=+0.8>
Agora que os valores ausentes foram preenchidos vamos mudar os dados para números inteiros.

```python
# Alterando o tipo dos dados
df['year_of_release'] = df['year_of_release'].astype('int')

# Exibindo informações
df.info()
```

3.2.3 Coluna <b>critic_score</b>.

    <p>
<font face='Calibri' color="#001A56" size=+0.8>
Analisaremos agora os valores ausentes da coluna <b>critic_score</b>.

<font face='Calibri' color="#001A56" size=+0.8>
Possuímos 8576 valores ausentes nessa coluna, mas será que a avaliação da crítica tem relação com alguma outra característica? é o que analisaremos agora.

Como podemos ver, inicialmente não há uma forte correlação com nenhuma outra caracteristica. Vamos tentar agrupar os dados por plataformas e por gênero para verificar se é possível inferir algo dessas caracteristicas agrupadas.

```python
# Agrupando por gênero e plataforma
df.pivot_table(index=['platform','genre'], values='critic_score', aggfunc=['mean','median', 'count']).head(10)
```

<font face='Calibri' color="#001A56" size=+0.8>
Notamos que há diferenças entre as avaliações quando agrupadas por plataformas e gêneros, porém dessa forma temos muito grupos pequenos, com poucas amostras, por isso não utilizaremos as médias ou medianas dos dados agrupados dessa forma.

Esses jogos que possuem valores ausentes não podem ser excluídos porque representam uma parcela muito grande do volume total de dados, quase 50%. Porém, apresentam uma correlação muito pequena com as outras informações, também não podemos inferir que todos os jogos do mesmo gênero ou mesma plataforma recebem as mesmas notas, por esses motivos iremos manter os valores ausentes.

<h4><font face='Calibri' color="#1E90B6" size=+1.8>
3.3.1 Coluna <b>total_sales</b>.
</font></h4>
    <p>
<font face='Calibri' color="#001A56" size=+0.8>
Iremos incluir uma nova coluna(<b>total_sales</b>) que irá conter a informação do total de vendas de cada jogo em todas as regiões.
    </font>

```python
#Função para calcular o total de vendas.
def venda_total(row):
    total_sales = row['na_sales'] + row['eu_sales'] + row['jp_sales'] + row['other_sales']
    return total_sales

#Testando a função.
venda_total(df.iloc[4])
```

```python
#Aplicando a função.
df['total_sales'] = df.apply(venda_total, axis= 1)

# Exibindo as primeiras linhas
df.head()
```

<a id='4.2'></a>
<h3><font face='Calibri' color="#013294" size=+2.0>
4.2 Principais plataformas.
</font></h3>
    
<p>
<font face='Calibri' color="#001A56" size=+0.8>
Agora analisaremos quais são as principais plataformas, avaliaremos sua popularidade segundo o total de vendas de cada uma delas.

```python
# Agrupando por plataforma
df.groupby('platform')['total_sales'].sum().sort_values(ascending=False)
```

```python
print('Média de vendas por plataformas: {:.2f}'.format(df.groupby('platform')['total_sales'].sum().mean()))
```

```python
# Plotando gráfico
df.groupby('platform')['total_sales'].sum().sort_values(ascending=False).plot(
    kind='bar', figsize=(12,6), color='#03588C', width=0.7, grid=False)

_ = plt.title('Venda total por plataforma', fontsize=16)
plt.ylabel('Total de vendas em milhões de dólares (USD)', fontsize=14)
plt.xlabel('Plataformas', fontsize=14)
sns.despine()
```

<font face='Calibri' color="#001A56" size=+0.8>
Acima temos o total de vendas de todas as plataformas, como há bastante diferença entre elas, iremos analisar apenas as plataformas que possuem mais de 287,5 milhões em vendas totais, que é a média de vendas das plataformas.

```python
# Agrupando os dados
sales = df.pivot_table(index='platform', values='total_sales', aggfunc='sum')

# Filtrando os dados
sales = sales[sales['total_sales'] > 287.5]

# Exibindo os dados
sales.head(10)
```

<font face='Calibri' color="#001A56" size=+0.8>
Conforme o critério que definimos, iremos analisar as 9 plataformas acima. Começaremos filtrando de nosso DataFrame principal apenas os jogos que pertencem à essas plataformas.

```python
# Filtrando dados
top_sales = df.query('platform in @sales.index') 

# Exibindo os dados
top_sales.head()
```

<font face='Calibri' color="#001A56" size=+0.8>
Agora criaremos uma tabela dinâmica com a contagem de jogos de cada plataforma por ano.

```python
# Lançamentos anuais por plataforma.
top_sales.pivot_table(
    index='year_of_release', columns='platform', values='name', aggfunc='count', fill_value='-')
```

```python
# Plotando gráfico
top_sales.pivot_table(
    index='year_of_release', columns='platform', values='total_sales', aggfunc='sum').plot(
    figsize=(12,6), grid=False, style='-o')

_ = plt.title('Venda total das plataformas ao longo do tempo', fontsize=16)
plt.ylabel('Total de vendas em milhões de dólares (USD)', fontsize=14)
plt.xlabel('Ano de Lançamento', fontsize=14)
plt.legend()
sns.despine()
```

<font face='Calibri' color="#001A56" size=+0.8>

```python
# Plotando gráfico
games.groupby('platform')['name'].count().sort_values(ascending=False).plot(
    kind='bar', figsize=(12, 6), color='#03588C', width=0.7, grid=False)

_ = plt.title('Quantidade de jogos do PS4 em outras plataformas', fontsize=16)
plt.ylabel('Quantidade de jogos', fontsize=14)
plt.xlabel('Plataformas', fontsize=14)
plt.xticks(rotation=0)
sns.despine()
```

A correlação entre as avaliações da crítica e as vendas nas plataformas que possuem maiores quantidade dos mesmo jogos que o PS4, são semelhantes à aquelas que vimos no próprio PS4.

6. Testanto Hipóteses.
</font></h2>
    <p>
<a id='6.1'></a>
<h3><font face='Calibri' color="#013294" size=+2.0>
6.1 Classificação dos usuário de XOne e PC.</font></h3></p>
    
<p>
<font face='Calibri' color="#001A56" size=+0.8>
Começaremos a testar algumas hipóteses, a primeira é que as classificações médias dos usuários das plataformas Xbox One e PC são as mesmas. Antes de realizar o ttest, vamos conmhecer as médias e a variância das classificações dos usuários de cada plataforma:

```python
# Filtrando os jogos da plataforma XOne.
data_xone = data[data['platform'] == 'XOne']

# Obtendo média e variância das avaliações dos usuários.
xone_user_score_mean = data_xone['user_score'].mean()
xone_user_score_var = np.var(data_xone['user_score'])

print('Avaliações dos uusuários XOne - média:', xone_user_score_mean)
print('Avaliações dos uusuários Xone - variância:', xone_user_score_var)
```

```python
# Filtrando os jogos da plataforma PC.
data_pc = data[data['platform'] == 'PC']

# Obtendo média e variância das avaliações dos usuários.
pc_user_score_mean = data_pc['user_score'].mean()
pc_user_score_var = np.var(data_pc['user_score'])

print('Avaliações dos uusuários PC - média:', pc_user_score_mean)
print('Avaliações dos uusuários PC - variância:', pc_user_score_var)
```

<font face='Calibri' color="#001A56" size=+0.8>
O primeiro teste de hipóteses que iremos fazer é referente a igualdade ou diferença entre as classificações médias dos usuários das plataformas Xbox One e PC, levando em consideração o teorema central do limite que nos diz que a distribuição deve ser normal, e a média é o valor proposto em H₀, temos as seguintes hipóteses:

<ul>
<li>Hipótese nula:</li><p>
    <ul>
    <li>H₀: As classificações médias dos usuários das plataformas Xbox One e PC são iguas.</li>
    </ul>
   
<p>       
<li>Hipótese alternativa:</li>
<ul><li>Ha: As classificações médias dos usuários das plataformas Xbox One e PC são diferentes.</li>
    </ul></ul>
    
Como as varianças das avaliações das plataformas em questão são diferentes utilizaremos o parametro <code>equal_var</code> como <b>False</b>. Definiremos o valor do limiar <i><b>alfa</b></i> como 0.05.

```python
# Definindo Alfa
alpha = 0.05

# Realizando teste
results = st.ttest_ind(data_xone['user_score'], data_pc['user_score'], equal_var = False, nan_policy='omit')

print('p-value:', results.pvalue)

if results.pvalue < alpha:
    print("Hipótese nula rejeitada! ")
else:
    print("Não é possível rejeitar a hipótese nula! ")
```

<font face='Calibri' color="#001A56" size=+0.8>
O resultado do ttest não nos permite rejeitar a hipotese nula, que ss classificações médias dos usuários das plataformas Xbox One e PC são iguas.

<a id='6.2'></a>
<h3><font face='Calibri' color="#013294" size=+2.0>
6.2 Classificações de usuários dos gêneros Action (ação) e Sports (esportes).</font></h3></p>
    
<p>
<font face='Calibri' color="#001A56" size=+0.8>
Agora testaremos a hipótese que as classificações médias dos usuários dos gêneros Action (ação) e Sports (esportes) são diferentes. Antes de realizar o ttest, vamos conmhecer as médias e a variância das classificações dos usuários de cada gênero:

```python
# Filtrando os jogos do gênero Ação.
data_action = data[data['genre'] == 'Action']

# Obtendo média e variância das avaliações dos usuários.
action_user_score_mean = data_action['user_score'].mean()
action_user_score_var = np.var(data_action['user_score'])

print('Avaliações dos uusuários do gênero Action - média:', action_user_score_mean)
print('Avaliações dos uusuários do gênero Action - variância:', action_user_score_var)
```

```python
# Filtrando os jogos do gênero Esportes.
data_sports = data[data['genre'] == 'Sports']

# Obtendo média e variância das avaliações dos usuários.
sports_user_score_mean = data_sports['user_score'].mean()
sports_user_score_var = np.var(data_sports['user_score'])

print('Avaliações dos uusuários do gênero Sports - média:', sports_user_score_mean)
print('Avaliações dos uusuários do gênero Sports - variância:', sports_user_score_var)
```

<font face='Calibri' color="#001A56" size=+0.8>
Essa hipótese que iremos testar é diz respeito a igualdade ou diferença entre as classificações médias dos usuários dos gêneros Action e Sports, mais uma vez levaremos em consideração o teorema central do limite que nos diz que a distribuição deve ser normal, e a média é o valor proposto em H₀, temos as seguintes hipóteses:
<ul>
<li>Hipótese nula:</li><p>
    <ul>
    <li>H₀: As classificações médias dos usuários dos gêneros Action e Sports são iguas.</li>
    </ul>
   
<p>       
<li>Hipótese alternativa:</li>
<ul><li>Ha: As classificações médias dos usuários dos gêneros Action e Sports são diferentes.</li>
    </ul></ul>

Como as varianças das avaliações dos gêneros são diferentes utilizaremos o parametro <code>equal_var</code> como <b>False</b>. Definiremos o valor do limiar <b><i>alfa</i></b> como 0.05.

```python
# Definindo alfa
alpha = 0.05

# Realizando teste
results = st.ttest_ind(data_action['user_score'], data_sports['user_score'], equal_var = False, nan_policy='omit')

print('p-value:', results.pvalue)

if results.pvalue < alpha:
    print("Hipótese nula rejeitada! ")
else:
    print("Não é possível rejeitar a hipótese nula! ")
```

<font face='Calibri' color="#001A56" size=+0.8>
Rejeitaremos a hipótese nula que as classificações médias dos usuários dos gêneros Action e Sports são iguas, pois o resultado do <i>p-value</i> foi superior ao <i>alfa</i> que definimos.</font>
