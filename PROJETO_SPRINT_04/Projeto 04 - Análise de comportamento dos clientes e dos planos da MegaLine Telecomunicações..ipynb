<font face='Calibri' color="#001A56" size=+0.8>
O gráfico acima reforça a idéia que o número de usuários aumentou no decorrer do tempo, esse aumento na quantidade de clientes pode variar em outras métricas que analisaremos. Caso a informação do aumento da quantidade de clientes por mês se confirme teremos que ter cuidado ao avaliar as métricas mensais.

```python
# Plotando gráfico
df_calls['duration'].hist(color='#191659', figsize=(12, 6), bins=30, grid=False)
_ = plt.title('Frequência de duração das chamadas', fontsize=16)
plt.xlabel('Duração de chamadas', fontsize=14)
plt.ylabel('Frequência', fontsize=14)
sns.despine()
```

<font face='Calibri' color="#001A56" size=+0.8>
De fato, podemos observar uma quantidade muito elevada de chamadas com 0.0 minutos de duração. O correto seria tentar entender o motivos desses registros, se é referente a chamadas perdidas, algum problema no imput dos dados ou chamadas extremamente curtas.

```python
# Contando chamandas de 0seg de duração
df_calls[df_calls['duration'] == 0.00].count()
```

```python
# Contando chamandas de 0.1
df_calls[df_calls['duration'] == 0.01].count()
```

```python
# Checando a porcetagem das chamadas de 0seg.
len(df_calls[df_calls['duration'] == 0.0])/len(df_calls)
```

<font face='Calibri' color="#001A56" size=+0.8>
Aproximadamente 20% das chamadas tiveram duração de 0.0 minutos. Não encontramos nenhuma informação que possa confirmar que essas chamadas possam ser chamadas perdidas. Também descarto a hipótese de chamadas extremamente rápidas, pois temos registros de chamadas com duração de 0.01 que corresponderia a chamadas com menos de 1 segundo.

Temos a orientação que a duração das chamadas devem ser arredondas para cima, como o arrendondamento de 0.0 é o próprio 0 esses registros não irão gerar um desconto do pacote de minutos do plano selecionado pelo cliente, como não causara impacto no faturamento que é a principal informação que nos interessa os dados serão mantidos.

5.1  Calculando o consumo de mensagens.
    </font></h3>
<p>    
<font face='Calibri' color="#001A56" size=+0.8>    
Criaremos uma coluna com a informação do valor adicional pago pelo usuário em caso de mensagens excedentes ao plano utilizado. Coluna <b>cost_mens</b>.

Começaremos criando uma função para realizar o cálculo.

```python
# Criando a função
def calculo_mensagens(row):
    if row['plan'] == 'ultimate':
        if row['messages'] > 1000:
            return (row['messages'] - 1000) * 0.01
        else:
            return 0.00
    elif row['plan'] == 'surf':
        if row['messages'] > 50:
            return (row['messages'] - 50) * 0.03
        else:
            return 0.00
        
# Testando a função
calculo_mensagens(data.iloc[0])
```

<font face='Calibri' color="#001A56" size=+0.8>
Função criada, com ela agora criaremos a coluna.

```python
# Criando coluna
data['cost_mens'] = data.apply(calculo_mensagens, axis=1)

# Exibindo as primeiras linhas
data.head()
```

<font face='Calibri' color="#001A56" size=+0.8>
Função utilizada e coluna criada.

<a id='5.2'></a>   
<h3><font face='Calibri' color="#013294" size=+2.0>
5.2  Calculando o consumo de internet.
    </font></h3>
<p>    
<font face='Calibri' color="#001A56" size=+0.8>    
Criaremos uma coluna com a informação do valor adicional pago pelo usuário em caso de uso de internet excedente ao plano utilizado. Coluna <b>cost_net</b>.

Começaremos criando uma função para realizar o cálculo.

```python
# Criando função
def calculo_internet(row):
    if row['plan'] == 'ultimate':
        if row['gb_used'] > 30:
            return (row['gb_used'] - 30) * 0.07
        else:
            return 0.00
    elif row['plan'] == 'surf':
        if row['gb_used'] > 15:
            return (row['gb_used'] - 15) * 0.10
        else:
            return 0.00
        
# Testando função
calculo_internet(data.iloc[0])
```

<font face='Calibri' color="#001A56" size=+0.8>
Criando a coluna <b>cost_net</b>.

```python
# Criando a coluna
data['cost_net'] = data.apply(calculo_internet, axis=1)

# Exibindo as primeiras linhas
data.head()
```

<font face='Calibri' color="#001A56" size=+0.8>
Função utilizada e coluna criada.

<h3><font face='Calibri' color="#013294" size=+2.0>
5.3  Calculando o consumo de chamadas
    </font></h3>
<p>    
<font face='Calibri' color="#001A56" size=+0.8>    
Criaremos uma coluna com a informação do valor adicional pago pelo usuário em caso de ligações excedente ao plano utilizado. Coluna <b>cost_calls</b>.

Começaremos criando uma função para realizar o cálculo.

```python
# Criando a função
def calculo_ligacoes(row):
    if row['plan'] == 'ultimate':
        if row['duration'] > 3000:
            return (row['duration'] - 3000) * 0.01
        else:
            return 0.00
    elif row['plan'] == 'surf':
        if row['duration'] > 500:
            return (row['duration'] - 500) * 0.03
        else:
            return 0.00
        
# Testando a função
calculo_ligacoes(data.iloc[0])
```

<font face='Calibri' color="#001A56" size=+0.8>
Criando a coluna <b>cost_calls</b>.

```python
# Criando a coluna
data['cost_calls'] = data.apply(calculo_ligacoes, axis=1)

# Exibindo as primeiras linhas
data.head()
```

Função utilizada e coluna criada.

<h3><font face='Calibri' color="#013294" size=+2.0>
5.4  Calculando o consumo total.
    </font></h3>
<p>    
<font face='Calibri' color="#001A56" size=+0.8>    
Criaremos uma coluna com a informação do valor total pago pelo usuário. Coluna total_cost.

Começaremos definindo os valores adicionais por excedente de consumo e depois criaremos uma função para realizar o cálculo.

```python
# Definindo variáveis com valor mensal do plano Surf e Ultimate.
surf_monthly = 20

ultimate_monthly = 70

# Criando função
def calculo_total(row):
    if row['plan'] == 'ultimate':
        return ultimate_monthly + row['cost_mens'] + row['cost_net'] + row['cost_calls']
    elif row['plan'] == 'surf':
        return surf_monthly + row['cost_mens'] + row['cost_net'] + row['cost_calls']
    
# Testando função
calculo_total(data.iloc[3])
```

Criaremos a coluna <b>total_cost</b>.

```python
# Criando coluna
data['total_cost'] = data.apply(calculo_total, axis=1)

# Exibindo as primeiras linhas
data.head()
```

```python
# Calculando limites para outliers
Q1_gb_used_surf = surf_users['gb_used'].quantile(0.25)
Q3_gb_used_surf = surf_users['gb_used'].quantile(0.75)
IQR_gb_used_surf = Q3_gb_used_surf - Q1_gb_used_surf
lim_inf_gb_used_surf = Q1_gb_used_surf - (1.5*IQR_gb_used_surf)
lim_sup_gb_used_surf = Q3_gb_used_surf + (1.5*IQR_gb_used_surf)

print('Outlier inferior gb_used:', lim_inf_gb_used_surf)
print('Outlier superior gb_used:', lim_sup_gb_used_surf)
```

<font face='Calibri' color="#001A56" size=+0.8>
Filtraremos os dados sem os valores atípicos.

```python
# Filtrando dados
surf_filter_gb = surf_users[surf_users['gb_used'] < lim_sup_gb_used_ult]

# Checando o tamanho da tabela
len(surf_filter_gb)
```

8.1 Hipótese 1
    </font></h3>
    
<a id='8.1.1'></a>
<h4><font face='Calibri' color="#1E90B6" size=+1.5>
8.1.1  Receita média dos planos.
</font></h4>
    <p>
<font face='Calibri' color="#001A56" size=+0.8>
    O primeiro teste de hipóteses que iremos fazer é referente a igualdade ou diferença entre as receitas médias dos usuários dos planos Ultimate e Surf, levando em consideração o teorema central do limite que nos diz que a distribuição deve ser normal, e a média é o valor proposto em H₀, temos as seguintes hipóteses:

Hipótese nula:

<i>H₀: A receita média dos usuários dos planos Ultimate e Surf são iguas.</i>

<p>
Hipótese alternativa:
    
<i>Ha: A receita média dos usuários dos planos Ultimate e Surf são diferentes.</i>
<p>    
Como as varianças das receitas dos planos são diferentes utilizaremos o parametro <b><i>equal_var</i></b> como <i>False</i>.

```python
# Teste de hipóteses
alpha = 0.05

results = st.ttest_ind(data_surf['total_cost'], data_ultimate['total_cost'], equal_var = False)

print('p-value:', results.pvalue)

if results.pvalue < alpha:
    print("Hipótese nula rejeitada! ")
else:
    print("Não é possível rejeitar a hipótese nula! ")
```

<font face='Calibri' color="#001A56" size=+0.8>
    Conforme o teste acima nos mostrou , não temos evidências que a receita média dos usuários dos planos Ultimate e Surf são iguais.
